Stock Market Crash Analysis

This notebook analyzes historical Sensex data to identify and visualize market crashes, drawdowns, and early-warning signals. The code cells below load the data, compute returns and drawdowns, and produce Plotly visualizations with improved styling and hover information.

### Import Required Libraries

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

### Load Dataset

```python
df = pd.read_csv('cleaned_sensex.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df.set_index('Date', inplace=True)

df.head()
```

### Calculate Daily Returns

```python
df['Daily_Return'] = df['Close'].pct_change() * 100
df.head()
```

### Identify Daily Crash Days

```python
CRASH_THRESHOLD_DAILY = -5

df['Crash_Daily'] = df['Daily_Return'] <= CRASH_THRESHOLD_DAILY
df[['Close', 'Daily_Return', 'Crash_Daily']].head(20)
```

```python
CRASH_THRESHOLD_DAILY = -5

df['Crash_Daily'] = df['Daily_Return'] <= CRASH_THRESHOLD_DAILY
df[['Close', 'Daily_Return', 'Crash_Daily']].tail(20)
```

```python
CRASH_THRESHOLD_DAILY = -5

# Crash column already hai
df['Crash_Daily'] = df['Daily_Return'] <= CRASH_THRESHOLD_DAILY

# Sirf crash days nikal lo
crash_days = df[df['Crash_Daily'] == True]
```

```python
crash_days = crash_days.copy()
crash_days['Year'] = crash_days.index.year
crash_days['Date'] = crash_days.index.date  # optional, just date
```

### Plot Closing Price with Daily Crashes

```python
import plotly.io as pio
print(pio.renderers)
```

```python
import plotly.io as pio
pio.renderers.default = "vscode"

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df.index,
    y=df['Close'],
    mode='lines',
    name='Closing Price'
))

fig.add_trace(go.Scatter(
    x=df.index[df['Crash_Daily']],
    y=df['Close'][df['Crash_Daily']],
    mode='markers',
    marker=dict(color='red'),
    name='Daily Crash'
))

fig.update_layout(
    title="Sensex Closing Price with Daily Crashes",
    xaxis_title="Date",
    yaxis_title="Index Value",
    xaxis_rangeslider_visible=True,
    template="plotly_white"
)

fig.show()
```

### Calculate Drawdowns

```python
df['Cumulative_Max'] = df['Close'].cummax()
df['Drawdown'] = (df['Close'] - df['Cumulative_Max']) / df['Cumulative_Max'] * 100

df[['Close', 'Cumulative_Max', 'Drawdown']].head()
```

### Plot Market Drawdown

```python
DRAWDOWN_THRESHOLD = -20
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df.index,
    y=df['Drawdown'],
    mode='lines',
    name='Drawdown (%)'
))

fig.add_hline(
    y=DRAWDOWN_THRESHOLD,
    line_dash='dash',
    line_color='red',
    annotation_text='Crash Threshold (-20%)'
)

fig.update_layout(
    title="Market Drawdown Over Time",
    xaxis_title="Date",
    yaxis_title="Drawdown (%)",
    xaxis_rangeslider_visible=True,
    template="plotly_white"
)

fig.show()
```

### Extract Severe Drawdown Dates

```python

crash_drawdowns = df[df['Drawdown'] <= DRAWDOWN_THRESHOLD]
crash_drawdowns[['Close', 'Drawdown']].head(10)
```

### Detect Crash Clusters

```python
crash_dates = crash_drawdowns.index

clusters = []
current_cluster = []

for date in crash_dates:
    if not current_cluster:
        current_cluster.append(date)
    elif (date - current_cluster[-1]).days <= 3:
        current_cluster.append(date)
    else:
        clusters.append(current_cluster)
        current_cluster = [date]

if current_cluster:
    clusters.append(current_cluster)

for i, cluster in enumerate(clusters[:20]):
    print(f"Cluster {i+1}: {cluster[0].date()}  {cluster[-1].date()} ({len(cluster)} days)")
```

### Zoom into First Crash Cluster

```python
cluster_start = clusters[0][0]
cluster_end = clusters[0][-1]

start_date = cluster_start - pd.Timedelta(days=30)
end_date   = cluster_end + pd.Timedelta(days=30)

zoom_df = df[(df.index >= start_date) & (df.index <= end_date)]
```

### Plot Zoomed Closing Price

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=zoom_df.index,
    y=zoom_df['Close'],
    mode='lines',
    name='Close'
))

fig.add_vrect(
    x0=cluster_start,
    x1=cluster_end,
    fillcolor='red',
    opacity=0.3,
    layer='below',
    line_width=0
)

fig.update_layout(
    title="Zoomed Crash Period: Closing Price",
    xaxis_rangeslider_visible=True,
    template="plotly_white"
)

fig.show()
```

### Define Crash Plot Functions

```python
def plot_crash_prices(start, end, title):
    # Safe slice using boolean mask
    zoom = df[(df.index >= start - pd.Timedelta(days=30)) & 
              (df.index <= end + pd.Timedelta(days=30))]
    
    fig = go.Figure()
    
    # Closing price line
    fig.add_trace(go.Scatter(x=zoom.index, y=zoom['Close'], mode='lines'))
    
    # Highlight crash cluster
    fig.add_vrect(x0=start, x1=end, fillcolor='red', opacity=0.3)
    
    fig.update_layout(
        title=title,
        xaxis_rangeslider_visible=True,
        template="plotly_white"
    )
    fig.show()
```

### Plot Major Crashes

```python
plot_crash_prices(pd.to_datetime("1997-11-12"), pd.to_datetime("1997-12-26"), "1997 Crash")
plot_crash_prices(pd.to_datetime("2008-08-18"), pd.to_datetime("2009-01-23"), "200809 Crash")
plot_crash_prices(pd.to_datetime("2020-03-16"), pd.to_datetime("2020-04-03"), "2020 COVID Crash")
```

### Simulate Synthetic 2027 Data

```python
np.random.seed(42)

dates = pd.bdate_range("2027-01-01", periods=250)

returns = np.concatenate([
    np.random.normal(0.0005, 0.01, 150),
    np.random.normal(-0.008, 0.025, 50),
    np.random.normal(0.0005, 0.01, 50)
])

prices = [30000]
for r in returns:
    prices.append(prices[-1] * (1 + r))

df_2027 = pd.DataFrame({
    'Close': prices[1:],
    'Daily_Return': returns * 100
}, index=dates)

df_2027.head()
```

### Rolling Indicators (Early Warning System)

```python
df_2027['Rolling_Mean_Return'] = df_2027['Daily_Return'].rolling(10).mean()
df_2027['Rolling_Volatility'] = df_2027['Daily_Return'].rolling(10).std()

df_2027['Warning'] = (
    (df_2027['Rolling_Mean_Return'] < -0.5) &
    (df_2027['Rolling_Volatility'] > 2)
)


df_2027[df_2027['Warning']].head(10)
```

### Plot Early Warning Signals

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_2027.index,
    y=df_2027['Close'],
    mode='lines',
    name='Close'
))

fig.add_trace(go.Scatter(
    x=df_2027.index[df_2027['Warning']],
    y=df_2027['Close'][df_2027['Warning']],
    mode='markers',
    marker=dict(color='red'),
    name='Early Warning'
))

fig.update_layout(
    title="Synthetic 2027 Market with Early Warning Signals",
    xaxis_rangeslider_visible=True,
    template="plotly_white"
)

fig.show()
```
